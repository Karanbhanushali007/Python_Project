## Q2.
In a residential locality, the mean size of the house is 2224 square feet and the median value of the house is 1500 square feet. What can you interpret about the skewness in the distribution of house size? Are there bigger or smaller houses in the residential locality?

```python
# Data Avaialble
mean_size = 2224
median_size = 1500

# Skewness
if mean_size > median_size:
    skewness_direction = "right-skewed (positively skewed)"
    interpretation = "There are some larger houses in the residential locality, which is pulling the mean size up."
elif mean_size < median_size:
    skewness_direction = "left-skewed (negatively skewed)"
    interpretation = "There are some smaller houses in the residential locality, which is pulling the mean size down."
else:
    skewness_direction = "not skewed"
    interpretation = "The distribution of house sizes is symmetrical with zero skewness."

# Results
print(f"Mean Size: {mean_size} square feet")
print(f"Median Size: {median_size} square feet")
print(f"Skewness Direction: {skewness_direction}")
print(f"Interpretation: {interpretation}")
```

During the survey, the ages of 80 patients infected by COVID and admitted to one of the city hospitals were recorded and the collected data is represented in the less than cumulative frequency distribution table.a. Which class interval has the highest frequency?

c. How many patients aged 45 years and above were admit

e. What is the median class interval of age? age? age?


```python
import pandas as pd

# Data
data = {'Age (in yrs)': ['5-15', '15-25', '25-35', '35-45', '45-55', '55-65'],
        'No.of Patients': [6, 11, 21, 23, 14, 5]}
df = pd.DataFrame(data)

# a. Which class interval has the highest frequency?
max_frequency = df['No.of Patients'].max()
max_interval = df[df['No.of Patients'] == max_frequency]['Age (in yrs)']
print(f"Class interval with the highest frequency: {max_interval.values[0]}")

# b. Which age was affected the least?
min_frequency = df['No.of Patients'].min()
min_interval = df[df['No.of Patients'] == min_frequency]['Age (in yrs)']
print(f"Age interval affected the least: {min_interval.values[0]}")

# c. How many patients aged 45 years and above were admitted?
index = df[df['Age (in yrs)'] == '45-55'].index[0]
patients_45_and_above = df.loc[index:, 'No.of Patients'].sum()
print(f"Number of patients aged 45 years and above: {patients_45_and_above}")

# d. Which is the modal class interval in the above dataset
modal_class_interval = df[df['No.of Patients'] == max_frequency]['Age (in yrs)']
print(f"Modal class interval: {modal_class_interval.values[0]}")

# e. What is the median class interval of age?
df['Cumulative Frequency'] = df['No.of Patients'].cumsum()
total_patients = df['No.of Patients'].sum()
median_position = total_patients / 2
median_class_interval = df[df['Cumulative Frequency'] >= median_position].iloc[0]['Age (in yrs)']
print(f"Median class interval: {median_class_interval}")
```

## Q6. 
Suppose you have been told to measure the average height of all the males on the earth. What would be your strategy for the same? Would the average height be a parameter or a statistic? Justify your answer.

```python
# Sample data
sample = np.array([162, 170, 165, 185, 180, 172, 178, 169, 177, 166])

# Average height
avg = np.mean(sample)

# Standard deviation 
std_dev = np.std(sample)

#  Result
print(f"Average height from the sample: {avg:.2f} cm")
print(f"Standard deviation: {std_dev:.2f} cm")

# Assuming a large enough sample and normal distribution, calculate a 95% confidence interval
n = len(sample)
confidence_interval = 1.96 * (std_dev / np.sqrt(n)) #Margin of Error
lower_bound = avg - confidence_interval
upper_bound = avg + confidence_interval

print(f"95% Confidence Interval: ({lower_bound:.2f} cm, {upper_bound:.2f} cm)")

# Print the explanation about parameter and statistic
print("""
Parameter: A parameter is a value that describes a characteristic of an entire population.
In this case, the average height of all males on Earth is a parameter because it describes a characteristic of the global male population.

Statistic: A statistic is a value that describes a characteristic of a sample taken from the population.
The average height computed from your sample is a statistic because it describes a characteristic of the sample rather than the entire population.
""")
```

## Q7.
Calculate the z score of the following numbers:
X = [4.5,6.2,7.3,9.1,10.4,11]
Z score (Standard Score) = (Observed Value - Mean) / Std Deviation


```python
df = [4.5, 6.2, 7.3, 9.1, 10.4, 11]

# Calculating the mean
df_mean = np.mean(df)
# print(df_mean)

# Calculating the standard deviation
df_std_dev = np.std(df)
# print(df_std_dev)

# Calculating the z-scores
z_scores = (df - df_mean) / df_std_dev
# print(z_scores)

print("Mean:",df_mean)
print("Standard Deviation:", df_std_dev)
print("z-score:", z_scores)
```

## Q20.
Using the z-score of the income variable, find out the number of observations outside the +-3σ.

```python
# Selecting the Income column
income = df_1['Income']

# Calculating the mean and standard deviation of the 'Income' column
mean_income = income.mean()
std_income = income.std()

# Calculate the z-scores
z_scores = (income - mean_income) / std_income
# print(z_scores)

# Determining the number of observations outside ±3σ
outliers = np.abs(z_scores) > 3 #Creating a boolean Series where True indicates that the z-score is greater than 3 or less than 3 
# outliers
num_outliers = outliers.sum()

#Result
print(f"Number of observations outside ±3σ: {num_outliers}")
```
